# Расчет доходности акций в портфеле

```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import pandas_datareader as pdr
import matplotlib.pyplot as plt
```

```python
#Загружаем инструменты в портфеле
df1=DataFrame([['18.03.2014','TSLA','3'],
['10.11.2014','ORCL','20'],
['13.11.2014','MSFT','5'],
['24.04.2015','DAI','5'],
['06.10.2015','GOOGL','1']],columns=['Date','Name','Count'],dtype='int64')
df1.Date=pd.to_datetime(df1.Date,format='%d.%m.%Y')
df1
```

```python
#Загружаем данные по котировкам.
close_px_all = pdr.get_data_yahoo(df1['Name'])['Adj Close']
#Ресэмпл делаем для того, чтобы в случае указания даты выходного дня графики все равно строились
close_px_all=close_px_all.resample('D').ffill()
close_px_all.head()
```

```python
#Строим графики
for name in df1['Name']:
    print('Инструмент: ',df1[df1.Name==name]['Name'].values[0])
    print('Дата покумки: ',df1[df1.Name==name]['Date'].values)
    price=close_px_all.loc[df1['Date'][df1.Name==name]][name]
    print('Цена покупки: ', price.values[0])
    print('Текущая цена: ',close_px_all.loc[close_px_all.index.max()][name])
    print('Куплено акций: ', df1[df1.Name==name]['Count'].values[0])
    print('Потрачено: ',price.values[0] * df1[df1.Name==name]['Count'].values[0])
    close_px_all[name].loc[df1[df1.Name==name]['Date'].values[0]:].plot()
    plt.show()
```
